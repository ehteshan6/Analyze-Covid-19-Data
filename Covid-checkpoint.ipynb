```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('latest Covid-19 India Status1 (2).xls')
df
```

```python
df.isnull().sum()
```

```python
d=df.sort_values(by='Active_Ratio',ascending=False)
d
```

```python
df.head(2)
```

```python
df.columns=['State', 'Total_Cases', 'Active', 'Deaths', 'Active_Ratio', 'Death_Ratio']
```

```python
df.head(2)
```

```python
sns.barplot(x='State',y='Active_Ratio',data=d.head(10),palette='Set1')
plt.xticks(rotation=90,fontsize=8);
```

```python
D=df.sort_values(by='Death_Ratio',ascending=False)
D.head(2)
```

```python
plt.figure(figsize=(20,10),dpi=500)

plt.subplot(2,2,1)
plt.title('Top 10 Active Cases per State')
sns.barplot(x='State', y='Active_Ratio',data=d.head(10),palette='Set1')
plt.xticks(rotation=90,fontsize=8)

plt.subplot(2,2,2)
plt.title('Top 10 States with Highest Death ratio')
sns.barplot(x='State', y='Death_Ratio',data=D.head(10),palette='Set1')
plt.xticks(rotation=90,fontsize=8)

plt.subplot(2,2,3)
plt.pie('Active_Ratio',labels='State',data=d.head(10),autopct="%0.2f%%")

plt.subplot(2,2,4)
plt.pie('Death_Ratio',labels='State',data=D.head(10),autopct="%0.2f%%");
```

```python
plt.figure(figsize=(10,4),dpi=500)

plt.plot(df['State'],df['Active'],marker='o',label='Active')
plt.plot(df['State'],df['Deaths'],marker='o',label='Deaths')
plt.legend()
plt.xticks(rotation=90, fontsize=8);
```

```python
sns.lineplot(x='State', y='Active',marker='o', data=df, palette='Set1')

sns.lineplot(x='State', y='Deaths',marker='o', data=df, palette='Set1');
```

```python
df['Recovered_Cases']=df['Total_Cases']-(df['Active']+df['Deaths'])
```

```python
df['Recovered_Ratio']=df['Recovered_Cases']/df['Total_Cases']*100
```
